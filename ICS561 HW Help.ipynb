```python
import qiskit as q
import math
from qiskit.visualization import plot_histogram, plot_bloch_multivector
qasm_sim= q.Aer.get_backend("qasm_simulator")
statevec_sim = q.Aer.get_backend("statevector_simulator")
print("done setup")

# This circuit to practice Qiskit (For HW5 Q3)
c = q.QuantumCircuit(3,3)   # create a circuit with 3 qubits and 3 cbits. The cbits are needed to store the measerued qubits values. 

#part (a)
c.rx(math.pi/4, 0)   # Rotate qbit-0 45 degrees on X
c.ry(math.pi/4, 1)   # Rorate qbit-1 45 degrees on Y


#Part (b)
c.h(0)   # apply H gate on qbit-0
c.h(1)   # apply H gate on qbit-1

#Part (c)
c.draw()


```

```python
# Part (d) Display Block Spheres
psi0 = q.execute(c,backend=statevec_sim).result().get_statevector()
plot_bloch_multivector(psi0)
```

```python
# Testing rotation and CNOT gates 
c.rx(math.pi/2, 0)   # Rotate qbit-0 90 degrees on X
c.ry(math.pi/4, 1)   # Rorate qbit-1 45 degrees on Y
c.cnot(0,2)  # apply CNot gate on qubits (0 and 1)
c.cnot(1,2)  # apply CNot gate on qubits (1 and 3)
c.rz(math.pi/4, 2)  # Rotate qbit-2 45 degrees on Z
c.h(2)  # Apply H gate again on qbit-2
c.draw()
```

```python
# Part (e) Measuring 
c.measure([0,1,2], [0,1,2])   # measure qubits [0,1,2] and store the results in cbit [0,1,2]
c.draw()
```

```python
# Show the states using Block sphere
psi0 = q.execute(c,backend=statevec_sim).result().get_statevector()
plot_bloch_multivector(psi0)
```
